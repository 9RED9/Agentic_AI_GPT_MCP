# 실습 문제 모범답안 - 06. Context Management

`06_context_management` 노트북의 실습 문제에 대한 모범답안 예시입니다.

### 문제 요약

1. **쇼핑몰 개인화 도우미 (RunContextWrapper 응용)**:
   ① `ShopContext(user_name, is_vip, cart)` 정의 →
   ④ 장바구니 요약 도구(`is_vip`이면 할인 안내 포함) →
   ② 같은 에이전트를 VIP/일반 두 컨텍스트로 실행해 개인화 확인

2. **감사 로그 (ToolContext 응용)**:
   도구를 `ToolContext`로 수정하고, 호출될 때마다 `audit_log`에
   `{"user": ..., "tool": ..., "args": ...}`를 기록한 뒤 출력

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
Model = "gpt-5.4-mini"
```

## 1. 쇼핑몰 개인화 도우미

본문의 의존성 주입 4단계를 그대로 적용합니다.
- ① `ShopContext` 정의
- ② `Agent[ShopContext]` + `Runner.run(..., context=...)`
- ③ SDK가 자동으로 래핑·주입
- ④ 도구에서 `wrapper.context`로 접근

```python
from dataclasses import dataclass
from agents import Agent, Runner, RunContextWrapper, function_tool

# ① Context 정의
@dataclass
class ShopContext:
    user_name: str
    is_vip: bool
    cart: list


# ④ 도구에서 접근 - 장바구니 정보를 컨텍스트에서 직접 꺼냄 (LLM에게 묻지 않음)
@function_tool
async def summarize_cart(wrapper: RunContextWrapper[ShopContext]) -> str:
    """현재 사용자의 장바구니를 요약한다."""
    ctx = wrapper.context
    items = ", ".join(ctx.cart)
    summary = f"{ctx.user_name}님의 장바구니 ({len(ctx.cart)}개): {items}"
    if ctx.is_vip:
        summary += " / VIP 10% 할인 적용 예정"
    return summary


# ② 타입 지정
shop_agent = Agent[ShopContext](
    name="쇼핑 도우미",
    instructions="쇼핑몰 도우미입니다. 도구로 장바구니를 확인해서 간결하게 답변하세요.",
    model=Model,
    tools=[summarize_cart],
)
```

```python
# ② 전달 - VIP 고객
vip_user = ShopContext(user_name="김영희", is_vip=True, cart=["노트북", "무선 마우스", "USB 허브"])
result = await Runner.run(shop_agent, "제 장바구니 요약해 주세요.", context=vip_user)
print("VIP :", result.final_output)

# ② 전달 - 일반 고객 (같은 에이전트, 같은 질문, 다른 컨텍스트)
normal_user = ShopContext(user_name="박철수", is_vip=False, cart=["키보드", "모니터 받침대"])
result = await Runner.run(shop_agent, "제 장바구니 요약해 주세요.", context=normal_user)
print("일반:", result.final_output)
```

## 2. 감사 로그 (ToolContext 응용)

도구 시그니처만 `RunContextWrapper` → `ToolContext`로 교체하면
`tool_name`, `tool_call_id`, `tool_arguments` 메타데이터를 추가로 사용할 수 있습니다.
도구가 호출될 때마다 **"누가 · 어떤 도구를 · 어떤 인자로"** 호출했는지 `audit_log`에 기록합니다.

```python
from agents.tool_context import ToolContext

audit_log = []  # 감사 로그 저장소 (실제 서비스에서는 DB나 로그 시스템)

# 시그니처만 ToolContext로 교체 - 나머지 로직은 1번과 동일
@function_tool
async def summarize_cart_audited(ctx: ToolContext[ShopContext]) -> str:
    """현재 사용자의 장바구니를 요약한다. (호출 내역을 감사 로그에 기록)"""
    # 감사 로그 기록: 누가 어떤 도구를 어떤 인자로 호출했는가
    audit_log.append({
        "user": ctx.context.user_name,   # 누가 (컨텍스트)
        "tool": ctx.tool_name,           # 어떤 도구를 (메타데이터)
        "args": ctx.tool_arguments,      # 어떤 인자로 (메타데이터)
    })

    c = ctx.context
    summary = f"{c.user_name}님의 장바구니 ({len(c.cart)}개): {', '.join(c.cart)}"
    if c.is_vip:
        summary += " / VIP 10% 할인 적용 예정"
    return summary


audit_agent = Agent[ShopContext](
    name="쇼핑 도우미",
    instructions="쇼핑몰 도우미입니다. 도구로 장바구니를 확인해서 간결하게 답변하세요.",
    model=Model,
    tools=[summarize_cart_audited],
)
```

```python
# 두 사용자가 각각 도구를 호출
result = await Runner.run(audit_agent, "제 장바구니 요약해 주세요.", context=vip_user)
print("VIP :", result.final_output)

result = await Runner.run(audit_agent, "제 장바구니 요약해 주세요.", context=normal_user)
print("일반:", result.final_output)

# 감사 로그 확인 - 누가 어떤 도구를 어떤 인자로 호출했는지
print("\n[감사 로그]")
for entry in audit_log:
    print(entry)
```

### 정리

| 문제 | 핵심 포인트 |
|------|------|
| 1. 쇼핑몰 개인화 도우미 | 4단계 그대로: ① `ShopContext` → ② `Agent[ShopContext]` + `context=` → ③ 자동 → ④ `wrapper.context` |
| 2. 감사 로그 | 시그니처만 `ToolContext`로 교체 → `tool_name` / `tool_arguments` 메타데이터로 호출 내역 기록 |

- 장바구니·VIP 여부 같은 사용자 정보가 **프롬프트에 노출되지 않고** 도구가 직접 꺼내 씁니다.
- 감사 로그는 `ToolContext`의 대표 활용 사례입니다 — 도구 공용 로깅, `tool_call_id` 기반 요청 추적 등으로 확장할 수 있습니다.
