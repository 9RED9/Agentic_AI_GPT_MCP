# GPT와 MCP로 구현하는 자율형 AI Agent 구축 - 통합 교재 목차

이 문서는 Part I ~ Part III와 **10_~14_** 폴더 순서를 반영한 강의 진행 순서입니다.

---

## 전체 흐름

1. **Part I**: LangChain 없이 OpenAI Agent SDK만으로 Agent → RAG → Function Calling → MCP 구현
2. **Part II**: MCP 구현에 필요한 LangChain만 학습 (10_LangChain_MCP 내 이론)
3. **Part III**: LangChain을 이용한 MCP 구현 (11_ → 12_ → 13_ → 14_ → Udemy 5)
