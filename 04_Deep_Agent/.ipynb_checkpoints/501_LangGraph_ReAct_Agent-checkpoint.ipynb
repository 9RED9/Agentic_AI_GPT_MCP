# 계산기 에이전트 만들기

이 예제에서는 **LangGraph의 Graph API** 또는 **Functional API** 를 사용하여
**계산기 에이전트(calculator agent)** 를 만드는 방법을 보여줍니다.

* **Graph API**
  에이전트를 **노드(node)** 와 **엣지(edge)** 로 구성된 **그래프 형태**로 정의하고 싶을 때 사용합니다.
  (즉, 여러 처리 단계를 시각적으로 구성하고 싶은 경우)

* **Functional API**
  에이전트를 **하나의 함수(function)** 로 정의하고 싶을 때 사용합니다.
  (즉, 단일 입력 → 단일 출력 구조를 선호하는 경우)

```python
from dotenv import load_dotenv

# API-KEY 읽어오기
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

# model = init_chat_model("gpt-5.4-mini", model_provider="openai")
model = init_chat_model("gemini-2.5-flash", model_provider="google_genai")
```

```python
from langchain_core.tools import tool

# 도구 정의 (LLM이 사용할 수 있는 계산 기능)
@tool
def multiply(a: int, b: int) -> int:
    """`a`와 `b`를 곱합니다.

    매개변수(Args):
        a: 첫 번째 정수
        b: 두 번째 정수
    """
    return a * b

@tool
def add(a: int, b: int) -> int:
    """`a`와 `b`를 더합니다.

    매개변수(Args):
        a: 첫 번째 정수
        b: 두 번째 정수
    """
    return a + b

@tool
def divide(a: int, b: int) -> float:
    """`a`를 `b`로 나눕니다.

    매개변수(Args):
        a: 첫 번째 정수
        b: 두 번째 정수
    """
    return a / b

tools = [add, multiply, divide]
tools_by_name = {tool.name: tool for tool in tools}
model_with_tools = model.bind_tools(tools)
model_with_tools
```

### **상태(State) 정의**

그래프의 **상태(state)** 는 에이전트 실행 도중 **메시지(message)** 와 **LLM 호출 횟수(LLM calls)** 를 저장하는 데 사용됩니다.  
즉, LangGraph의 상태는 **에이전트가 동작하는 동안 유지되는 데이터 컨테이너** 역할을 하며,
대화의 흐름과 내부 처리 정보를 추적할 수 있게 합니다.

```python
from langchain.messages import AnyMessage
from typing_extensions import TypedDict, Annotated
import operator


class MessagesState(TypedDict):
    messages: Annotated[list[AnyMessage], operator.add]
    llm_calls: int
```

### **모델 노드(Model Node) 정의**

모델 노드는 **LLM(대형 언어 모델)** 을 호출하고, **도구(tool)** 를 호출해야 할지 여부를 결정하는 역할을 합니다.  
즉, 이 노드는 모델이 입력된 메시지를 바탕으로
“직접 대답할지” 혹은 “도구(예: 계산기, 검색기 등)를 사용해야 할지” 판단하는 **의사결정 지점**입니다.

```python
from langchain.messages import SystemMessage

def llm_call(state: dict):
    """LLM이 도구를 호출할지 여부를 결정합니다."""

    return {
        "messages": [
            model_with_tools.invoke(
                [
                    SystemMessage(
                        content="당신은 주어진 입력값에 대해 산술 연산을 수행하도록 설계된 친절한 도우미입니다."
                    )
                ]
                + state["messages"]  # 기존 대화 상태(state)에 포함된 메시지들을 모델에 전달
            )
        ],
        # LLM 호출 횟수를 1 증가시킴
        "llm_calls": state.get('llm_calls', 0) + 1
    }
```

### **도구 노드 정의 (Define tool node)**

도구 노드는 **등록된 도구(tools)** 를 실제로 호출하고, 그 **결과(result)** 를 반환하는 역할을 합니다.  

즉, LLM이 “도구를 사용해야 한다”고 판단했을 때 이 노드가 해당 도구를 실행하여 결과를 생성하고, 그 결과를 다시 모델(LLM)에게 전달합니다.

```python
from langchain.messages import ToolMessage

def tool_node(state: dict):
    """도구(tool)를 실제로 호출하고 결과를 반환합니다."""

    result = []
    # 마지막 메시지에 포함된 tool_calls(도구 호출 요청)를 순회
    for tool_call in state["messages"][-1].tool_calls:
        # 호출된 도구 이름에 해당하는 실제 도구 객체를 가져옴
        tool = tools_by_name[tool_call["name"]]
        # 해당 도구를 실행하고 결과를 observation으로 저장
        observation = tool.invoke(tool_call["args"])
        # 도구 실행 결과를 ToolMessage 형태로 변환 (tool_call_id는 LLM이 생성한 호출 ID와 일치해야 함)
        result.append(ToolMessage(content=observation, tool_call_id=tool_call["id"]))
    # 실행된 도구 결과를 messages 리스트 형태로 반환
    return {"messages": result}
```

### **엔드 로직 정의 (Define end logic)**

조건부 엣지 함수(conditional edge function)는 LLM이 도구 호출을 수행했는지 여부에 따라 **도구 노드(tool node)** 로 이동할지,
혹은 **프로세스를 종료(end)** 할지를 결정합니다.

```python
from typing import Literal
from langgraph.graph import StateGraph, START, END

def should_continue(state: MessagesState) -> Literal["tool_node", END]:
    """LLM이 도구 호출을 했는지 여부에 따라 루프를 계속할지, 중단할지를 결정합니다."""

    messages = state["messages"]
    last_message = messages[-1]  # 마지막 메시지를 가져옴

    # 만약 LLM이 도구 호출을 수행했다면 → 도구 노드로 이동하여 실행
    if last_message.tool_calls:
        return "tool_node"

    # 그렇지 않다면 → 루프를 종료하고 최종 응답을 반환
    return END
```

### **에이전트 구축 및 컴파일(Build and compile the agent)**

에이전트는 **`StateGraph` 클래스**를 사용하여 구축되고, **`compile()` 메서드**를 통해 실행 가능한 형태로 컴파일됩니다.

```python
# 워크플로우 빌드
agent_builder = StateGraph(MessagesState)

# 노드 추가
agent_builder.add_node("llm_call", llm_call)      # LLM 호출 노드
agent_builder.add_node("tool_node", tool_node)    # 도구 실행 노드

# 노드 간 연결(엣지) 추가
agent_builder.add_edge(START, "llm_call")  # 시작 → LLM 호출
agent_builder.add_conditional_edges(
    "llm_call",
    should_continue,                       # 조건 함수로 다음 경로 결정
    ["tool_node", END]                     # 도구 실행 또는 종료
)
agent_builder.add_edge("tool_node", "llm_call")   # 도구 실행 후 다시 LLM 호출

# 에이전트 컴파일 (실행 가능한 형태로 변환)
agent = agent_builder.compile()

# 에이전트 구조 시각화
from IPython.display import Image, display
display(Image(agent.get_graph(xray=True).draw_mermaid_png()))

# 에이전트 실행 (예시)
from langchain.messages import HumanMessage
messages = [HumanMessage(content="3 더하기 4를 계산해줘.")]
messages = agent.invoke({"messages": messages})
for m in messages["messages"]:
    m.pretty_print()
```
